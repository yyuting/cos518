```python
import numpy as np
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#This cell plots time taken by HogWild, RR, and Sync SGD to run 200 epochs 
#on synthetic dataset as a function of number of threads (10k, 0.2)

nthreads = np.array([1,2,4,8,10])
hogwild = np.array([1/154.27, 1/83.44, 1/55.19, 1/41.32, 1/37.79]) #time take by hogwild for different nthreads
rr = np.array([1/170.72048354, 1/138.30550861, 1/163.1898334,  1/183.5105712, 1/179.9780519]) #time take by RR for different nthreads
sync = np.array([1/157.08429885,  1/66.82344103,  1/45.06686568,  1/34.5469954,   1/31.06905723])


hogwild = hogwild*154.27
rr = rr*170.72048354
sync = sync*157.08429885

sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 4, 'lines.markersize': 10})

df = pd.DataFrame({
    'nthreads': nthreads, 
    'async': hogwild,
    'sync': sync,
    'RR': rr})
fig, ax = plt.subplots()
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.set_palette(flatui)
sns.lineplot(x='nthreads', y='value', hue='variable', style='variable', markers=True, dashes=False,
             data=pd.melt(df, ['nthreads']))
plt.xlabel(xlabel='Number of threads', size=20)
plt.ylabel(ylabel='Speedup', size=20)

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:], labels=labels[1:])
```
